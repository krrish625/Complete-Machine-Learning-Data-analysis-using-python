```python
class Marks:
   def __init__(self,number):
    self.n = number


   def factorial_while_loop( self):
    if self.n < 0:
        return "Factorial is not defined for negative numbers"
    elif self.n == 0:
        return 1
    else:
        result = 1
        i = 1
        # Loop as long as the counter is less than or equal to n
        while i <= self.n:
            result *= i
            i += 1  # Increment the counter
        return result




fact = Marks(5)
print("Factorial is:", fact.factorial_while_loop())

```

```python
#whithout inheritance
class parent:
  def p_method(self):
    print("this is parent method")

class child:
  def c_method(self):
    print("this is child method")

p = parent()
p.p_method()


c =child()
c.c_method()
```

```python
class parent:   #using inheritance
  def p_method(self):
    print("this is parent method")

class child(parent):
  def c_method(self):
    print("this is child method")


c =child()
c.p_method()
c.c_method()
```

```python
class father:  #parent class
  def land(self):
    print("this is father land ")

class child(father):   #child class
  def child_land(self):
    print("this is child land")

c = child()
c.land()
c.child_land()
```

```python
#multiple inheritance --> multiple parents , single child
```

```python
class parent1:   #using multiple  inheritance
  def p1_method(self):
    print("this is parent 1 method")

class parent2:
  def p2_method(self):
    print("this is parent 2 method")

class parent3:
  def p3_method(self):
    print("this is parent 3 method")


class child(parent1,parent2,parent3):
  def c_method(self):
    print("this is child method")


c =child()
c.p1_method()
c.p2_method()
c.p3_method()
c.c_method()
```

```python
#multilevel ---> grandfather , father , child
```

```python
class grandfather:    #multilevel inheritance
  def G_land(self):
    print("this is  grandfather land")

class father(grandfather):
  def F_land(self):
    print("this is father land")

class child(father):
  def C_land(self):
    print("this is child land")


c =child()
c.G_land()
c.F_land()
c.C_land()
```

```python
class parent:
  def parentmethod(self):
    print("this is parent method")

class child1(parent):
  def child1method(self):
    print("this is child 1 method")

class child2(parent):
  def child2method(self):
    print("this is child 2 method")

class child3(parent):
  def child3method(self):
    print("this is child 3 method")

c3 = child3()
c2 = child2()
c1 = child1()
c3.parentmethod()
c1.child1method()
c2.child2method()
c3.child3method()
c2.parentmethod()
c1.parentmethod()
```

```python
class grandparent:
  def legacy(self):
    print("grand parent")
class parent(grandparent):
  def parent(self):
    print("parent method")
class mother:
  def mother(self):
    print("mother method")
class child(mother,parent):
  def child(self):
    print("child method")

c = child()
c.legacy()
c.parent()
c.mother()
c.child()
```

```python
from abc import ABC,abstractmethod
class Fooddelivery(ABC):
  @abstractmethod
  def place_order(self,item):
    pass

  @abstractmethod
  def track_order(self,item):
    pass

class swiggi(Fooddelivery):
  def place_order(self, item):
    print("order placed for "+item)
    print("connecting with the restorent")
    print("assigning delivery partner")

  def track_order(self, item):
    print("your order is being prepared")
    print("delivery partner is on the way")

s = swiggi()
s.place_order("pizza")
s.track_order
```

```python
from abc import ABC,abstractmethod
class shape(ABC):
  @abstractmethod
  def area(self):
    pass

class circle(shape):
  def __init__(self,radius):
    self.radius=radius

  def area(self):
    print("area of circlr is :", 3.14*self.radius*self.radius)


class retangle(shape):
  def __init__(self,length,width):
    self.length=length
    self.width=width

  def area(self):
    print("area od rectangle is :",self.length*self.width)

class square(shape):
  def __init__(self,side):
    self.side = side

  def area(self):
    print("area of square is :",self.side*self.side)


  c1=circle(5)
  c1.area()

```

```python
r1=retangle(10,5)
r1.area()

s1=square(3)
s1.area()
```

**data selection and filtering in pandas**

```python
df
```

```python
# select those values where age value >20

df[df['Age']>20]
```

```python
# select rows where age >=30 and city is indor

filter_data = df[(df['Age']>=30) & (df['City']=='indor')]
filter_data
```

```python
# select rows where age <30 or city is dehli

filter_data = df[(df['Age']<30) | (df['City']=='dehli')]
filter_data
```

```python
df
```

```python
# select tha data where age is less than 30 but more than 25

filter = df[(df['Age']>25) & (df['Age']<30)]
filter
```
